From the ReadmeTwins file we notice the following about the data:
- X - includes 50 covariates for the twin pair. _0 and _1 denote differences in sex and birth order. The covariates pertain to parent characteristics, both mother and father.
- Y - mortality outcomes for twins, given by `mort_0` and `mort_1` respectively
- T - birth weights in grams for both twins

```python
for df in [df_X,df_Y,df_T]:
    print(str(df.info()) + "\n\n\n" + "=====================================================")
```

An interesting problem: can we predict what causally influences mortality within the twins?

To do this, we can apply linear regression here.

First, create a df that contains mortality and regressors.

```python
df_X.shape
```

```python
df_Y.shape
```

```python
mort_df = pd.concat([df_X,df_T,df_Y],axis=1)
mort_df
```

```python
mort_df.drop(["Unnamed: 0", "Unnamed: 0.1"],axis=1,inplace=True)
mort_df
```

Can we get the number of mortalities for the lower weighted vs. higher weighted twins?

```python
mort_df.shape[0]
```

```python
# Both twins survive
mort_df.loc[(mort_df['mort_0']==0) & (mort_df['mort_1']==0),["mort_0","mort_1"]].shape[0]
```

```python
# The second twin survives
mort_df.loc[(mort_df['mort_0']==1) & (mort_df['mort_1']==0),["mort_0","mort_1"]].shape[0]
```

```python
# The first twin survives
mort_df.loc[(mort_df['mort_0']==0) & (mort_df['mort_1']==1),["mort_0","mort_1"]].shape[0]
```

```python
# Neither survives
mort_df.loc[(mort_df['mort_0']==1) & (mort_df['mort_1']==1),["mort_0","mort_1"]].shape[0]
```

What would we expect to use as the 'treatment' in this case? What about birth weights? This is given by `dbirwt_0` and `dbirwt_1`.

What we might do is to create a new column with a difference in weight, and use that as the treatment.

```python
mort_df['dbirwt_diff'] = mort_df['dbirwt_1'] - mort_df['dbirwt_0']
mort_df.drop(['dbirwt_1','dbirwt_0'],axis=1,inplace=True)
```

What can we do given this information?

Combine these into averages. If we take the average number of twins that survive for each observation, we can see how likely it is for both to survive

```python
# This is what I tried previously
# mort_col = (mort_df["mort_0"].astype("bool") | mort_df["mort_1"].astype("bool")).astype("float")
```

```python
mort_col = (mort_df["mort_0"] + mort_df["mort_1"])/2
```

```python
mort_col.value_counts()
```

```python
mort_df = mort_df.drop(["mort_0","mort_1"],axis=1)
```

```python
# Combine into df
mort_df['mort_col'] = mort_col
```

```python
mort_df
```

Try this model:
    $$mort_i=\beta_0 + \beta_1 * diff_i + u_i$$

```python
model_1 = smf.ols('mort_col ~ dbirwt_diff',data=mort_df).fit()
model_1.summary().tables[1]
```

```python
mort_df['mort_col'].describe()
```

```python
mort_df['dbirwt_diff'].describe()
```

What can be said about this?
- Per gram increase in weight, the probability of surviving increases by 0.000051, or .0051%
    - Thus, for 284 grams (the mean) for instance, the probability of surviving increases by 1.45% 

Lets try another model - this time, lets include all the variables.

```python
mort_df
```

```python
controls = mort_df.drop(['dbirwt_diff','mort_col'],axis=1)
```

```python
model_2 = smf.ols('mort_col ~  dbirwt_diff+' + '+'.join(controls), data=mort_df).fit()
model_2.summary().tables[1]
```
